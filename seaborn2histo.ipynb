```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

## histograme

```python
df = sns.load_dataset("penguins") 
df
```

```python
sns.displot(df["flipper_length_mm"] )
plt.show()
```

```python
sns.displot(df["flipper_length_mm"] , bins=[170,180,190,200,210,220,230,240])  # it will arrange your data in systimatic way 
plt.show()
```

```python
sns.displot(df["flipper_length_mm"] , bins=[170,180,190,200,210,220,230,240] , kde=True)  # on bar you have a line 
plt.show()
```

```python
sns.displot(df["flipper_length_mm"] , bins=[170,180,190,200,210,220,230,240] , kde=True ,rug=True)  # on x axis  you have a line 
plt.show()
```

```python
sns.displot(df["flipper_length_mm"] , bins=[170,180,190,200,210,220,230,240] , kde=True ,rug=True , color="g")  # on x axis  you have a line 
plt.show()
```

```python
sns.displot(df["flipper_length_mm"]  , kde=True ,rug=True , color="g", log_scale=True)  # on x axis  you have a line 
plt.show()
```
