```python
import seaborn as sns
```

```python
tips.head()
```

Scatterplot

```python
sns.relplot(tips , x = "total_bill" , y ="tip" , kind="scatter")
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg.head()
```

```python
sns.relplot(mpg , x = "horsepower" , y = "mpg" , kind="scatter")
```

Hue

```python
sns.relplot(tips , x = "total_bill" , y ="tip" , kind="scatter" , hue = "smoker")
```

Rows and cols

```python
sns.relplot(tips , x = "total_bill" , y ="tip" , kind="scatter" , row = "smoker")
```

```python
sns.relplot(tips , x = "total_bill" , y ="tip" , kind="scatter" , col = "smoker")
```

```python
sns.relplot(tips , x = "total_bill" , y ="tip" , kind="line")
```

# Categorical Plots

```python
sns.relplot(tips , x = "smoker" , y ="total_bill")
```

```python
sns.catplot(tips , x = "smoker" , y = "total_bill")
```

```python
sns.catplot(tips , x = "smoker" , y = "total_bill" , kind = "swarm")
```
