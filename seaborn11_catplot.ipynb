```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
sns.catplot(x="day", y="size", data=df)

plt.title("catplot Plot Example")
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df , hue="sex")

plt.title("catplot Plot Example")
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df , hue="sex" ,palette="Oranges" ,height=7)

plt.title("catplot Plot Example")
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df , hue="sex" ,palette="Oranges" ,height=7 , kind="bar")   # imp kind= 

plt.title("catplot Plot Example")
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df , hue="sex" ,palette="Oranges" ,height=7 , kind="point")   # imp kind= 

plt.title("catplot Plot Example")
plt.show()
```

```python
sns.catplot(x="day", y="size", data=df , hue="sex" ,palette="Oranges" ,height=7 , kind="boxen")   # imp kind= 

plt.title("catplot Plot Example")
plt.show()
```
