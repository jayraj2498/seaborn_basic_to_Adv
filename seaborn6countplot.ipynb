```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
sns.countplot(x="sex"  , data=df)
plt.show()
```

```python
sns.countplot(x="sex"  , data=df , hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex"  , data=df , hue="smoker")   # for horizontally 
plt.show()
```

```python
sns.countplot(x="sex"  , data=df , hue="smoker" , color="r")
plt.show()
```

```python
sns.countplot(x="sex"  , data=df , hue="smoker" , color="r" , saturation=0.3)
plt.show()
```

```python
sns.countplot(x="day"  , data=df ,hue="time" , palette="BuGn")
plt.show()
```
