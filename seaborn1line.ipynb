```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
var = [1,2,3,4,5] 
var1= [2,3,6,5,6] 

data=pd.DataFrame({"var":var ,"var1":var1}) 

sns.lineplot( x= var , y = var1  , data = data ) 
plt.show()
```

```python
plt.plot(var,var1) 
plt.show()
```

```python
df = sns.load_dataset("penguins") 
df
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df) 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex")  # wrt to sex column 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" ,size=20)  # wrt to sex column 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" , style="sex")  # wrt to sex column 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" , style="sex" , palette="Accent")  # to change color
plt.show()
```

```python
# give markers in plot 
```

```python
df = sns.load_dataset("penguins").head(30)
df
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" , style="sex" , palette="Accent" ,
    markers= ["o" , "<"])  # to selesct markers 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" , style="sex" , palette="Accent" ,
    markers= ["o" , "<"] , dashes=False)  # to change  dash line 
plt.show()
```

```python
sns.lineplot(x='bill_length_mm' , y='flipper_length_mm'  , data=df , hue="sex" , style="sex" , palette="Accent" ,
    markers= ["o" , "<"] , dashes=False , legend=False)  # to remove legend  
plt.title("seaborn_chart")
plt.show()
```
