```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("penguins") 
df
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df) 

plt.show()
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df  , hue='sex') 

plt.show()
```

```python
df1=sns.load_dataset("penguins").head(40)
df1
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex") 

plt.show()
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex" , size="sex") 

plt.show()
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex" , size="sex",
                sizes=(100,10)) # according to  us 

plt.show()
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex" , size="sex",
                 palette="Accent") # to change the color  

plt.show()
```

```python
sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex" , size="sex",
                 palette="Accent" , alpha=0.3) # light , dark color  

plt.show()
```

```python
dic={"Male":"o" , "Female":"*"}

sns.scatterplot(x='bill_length_mm' , y='bill_depth_mm' , data=df1  , hue='sex' , style="sex" , size="sex",
                 palette="Accent" ,sizes=(100,50), markers=dic) # 

plt.show()
```
