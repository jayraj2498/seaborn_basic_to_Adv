```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
sns.set(style="whitegrid") 

sns.boxplot(x="day" ,y="total_bill" , data=df) 

plt.show()
```

```python
sns.set(style="whitegrid") 

sns.boxplot(x="day" ,y="total_bill" , data=df , hue="sex") 

plt.show()
```

```python
sns.set(style="whitegrid") 

sns.boxplot(x="day" ,y="total_bill" , data=df , hue="sex" ,color="g") 

plt.show()
```

```python
sns.set(style="whitegrid") 

sns.boxplot(x="day" ,y="total_bill" , data=df , hue="sex" ,color="g" , order=["Sun" ,"Sat","Thur","Fri"]) # order 

plt.show()
```

```python
sns.set(style="whitegrid") 

sns.boxplot(x="day" ,y="total_bill" , data=df , hue="sex" ,color="g" , order=["Sun" ,"Sat","Thur","Fri"] , showmeans=True)  

plt.show()
```
