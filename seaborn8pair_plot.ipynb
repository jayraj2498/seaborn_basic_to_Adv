```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
sns.pairplot(df) 
plt.show()
```

```python
sns.pairplot(df , hue="sex") 
plt.show()
```

```python
sns.pairplot(df , vars=["total_bill" , "tip"], hue="sex")   # pair plot wrt to col 
plt.show()
```

```python
sns.pairplot(df , vars=["total_bill" , "tip"], hue="sex" ,palette="BuGn")   # pair plot wrt to col 
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" ,palette="BuGn" , x_vars=["total_bill" , "tip"])  
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" ,palette="BuGn" , y_vars=["total_bill" , "tip"])  
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" , kind="reg")  
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" , kind="kde")  
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" , kind="kde" , diag_kind="hist")  
plt.show()
```

```python
sns.pairplot(df ,  hue="sex" , markers=["*" ,"<"])  
plt.show()
```
