```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("penguins") 
df
```

```python
sns.barplot(x=df.island , y=df.bill_length_mm)

plt.title("sns plot")
plt.show()
# plt.title("sns plot")
```

```python
#or 

sns.barplot(x='island' , y='bill_length_mm' , data=df)

plt.title("sns plot")
plt.show()
```

```python
# partition  wrt gendfer 

sns.barplot(x='flipper_length_mm' , y='bill_length_mm' , data=df , hue='sex')

plt.title("sns plot")
plt.show()
```

```python
# change the order 
# partition  wrt gender 

order1=['Dream','Biscoe','Torgersen']

sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , order=order1)

plt.title("sns plot")
plt.show()
```

```python

order1=['Dream','Biscoe','Torgersen']

sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , order=order1 , hue_order=["Female","Male"])  # hue order 

plt.title("sns plot")
plt.show()
```

```python
sns.barplot(x='flipper_length_mm' , y='bill_length_mm' , data=df , orient="v") 

plt.title("sns plot")
plt.show()
```

```python
sns.barplot(x='flipper_length_mm' , y='bill_length_mm' , data=df , orient="h" )    # horizontally 

plt.title("sns plot")
plt.show()
```

```python
#or 

sns.barplot(x='island' , y='bill_length_mm' , data=df , color="g")

plt.title("sns plot")
plt.show()
```

```python


sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , hue_order=["Female","Male"] , palette="icefire")  # color template

plt.title("sns plot")
plt.show()
```

```python


sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , hue_order=["Female","Male"] ,saturation=0.1)  # dark or light theme 

plt.title("sns plot")
plt.show()
```

```python


sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , hue_order=["Female","Male"] ,saturation=1)  # dark or light theme 

plt.title("sns plot")
plt.show()
```

```python


sns.barplot(x='island' , y='bill_length_mm' , data=df , hue='sex' , hue_order=["Female","Male"] ,saturation=1,alpha=0.3)  # dark or light theme 

plt.title("sns plot")
plt.show()
```
