```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
# it is same as box plot it works on categorical data   
```

```python
df=sns.load_dataset("tips") 
df
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df) 
plt.show()
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df , hue="time") 
plt.show()
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df , hue="time" , palette="Dark2") 
plt.show()
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df , hue="time" , palette="Dark2" , saturation=0.2) 
plt.show()
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df , hue="time" ,linewidth=2 , color="y") 
plt.show()
```

```python
sns.violinplot(x="day" , y="total_bill" , data=df , hue="sex") 
plt.show()
```
