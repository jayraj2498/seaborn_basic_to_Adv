```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
df1=sns.FacetGrid(df) 

df1.map(plt.scatter , "total_bill" ,"tip") 

plt.show()
```

## to make multiplot

```python
df1=sns.FacetGrid(df , col="sex") 

df1.map(plt.scatter , "total_bill" ,"tip") 

plt.show()
```

```python
df1=sns.FacetGrid(df , col="sex" ,hue="day") 

df1.map(plt.scatter , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="day" ,hue="sex") 

df1.map(plt.scatter , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="day" ,hue="sex") 

df1.map(plt.bar , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="sex" ,hue="day") 

df1.map(plt.bar , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="sex" ,hue="day" , palette="Accent") 

df1.map(plt.bar , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="day" ,hue="sex" , palette="Accent") 

df1.map(plt.bar , "total_bill" ,"tip").add_legend()

plt.show()
```

```python
df1=sns.FacetGrid(df , col="day" ,hue="sex" , palette="Accent") 

df1.map(plt.bar , "total_bill" ,"tip" ,edgecolor="r").add_legend()

plt.show()
```
