```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df=sns.load_dataset("tips") 
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.stripplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1 , edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1 , edgecolor="r" , jitter=10)  # to expand graph
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1 , edgecolor="r" ,size=10)  # to expand graph
plt.show()
```

```python
m={"Male":"O" , "Female":"<"}

sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1 , edgecolor="r"
    ,size=10 , marker=m)  # to expand graph
plt.show()
```

```python


sns.stripplot(x="day", y="total_bill", data=df , hue="sex" , palette="cool" , linewidth=1 , edgecolor="r"
    ,size=10 , alpha=0.3)  # to expand graph
plt.show()
```
